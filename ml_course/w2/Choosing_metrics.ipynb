```python
import pandas as pd
import numpy as np
from sklearn import neighbors
from sklearn import datasets, preprocessing, cross_validation
```

```python
boston = datasets.load_boston()
boston_features, boston_target = pd.DataFrame(preprocessing.scale(boston.data)), boston.target
boston_features.shape[0]
```

```python
k_folds = cross_validation.KFold(boston_features.shape[0], n_folds = 5, shuffle = True, random_state = 42)
```

```python
p_space = np.linspace(1, 10, 200)
scores = [(np.mean(cross_validation.cross_val_score(neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance', p=param), boston_features, boston_target, cv = k_folds, scoring = 'mean_squared_error')), param) for param in p_space]
best_score, best_p = max(scores)

print best_p, best_score

```
