```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.shape
```

```python
titanic.head()
```

```python
drop_titanic = titanic.drop(['SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked', 'PassengerId', 'Name'], axis = 1)
```

```python
drop_titanic.head()
```

```python
drop_titanic = drop_titanic.dropna()
```

```python
import numpy as np 
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=241)
```

```python
target = drop_titanic['Survived']
drop_titanic.drop(['Survived'], axis= 1, inplace = True)
```

```python
sex_binary = (drop_titanic['Sex']=='male').astype(int)
drop_titanic['Sex'] = sex_binary
drop_titanic
```

```python
clf.fit(drop_titanic, target)
```

```python
clf.feature_importances_
```
